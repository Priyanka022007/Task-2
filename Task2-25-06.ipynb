```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Remove Duplicates
ds.drop_duplicates(inplace=True)
```

```python
#To get columns only
ds.columns
```

# Handling Price

```python
#convert price into numeric format
ds["Price"]=pd.to_numeric(ds["Price"],errors='coerce')
ds.info()
```

```python
#To check Price has unique value have or not
ds["Price"].unique()
```

```python
#To fill data in column Price
x=ds["Price"].median()
ds["Price"].fillna(x,inplace=True)
ds.info()
ds
```

# Handling Quantity

```python
#To convert quantity  in nummeric form
ds["Quantity"]=pd.to_numeric(ds["Quantity"],errors='coerce')
ds.info()
```

```python
#To fill data in column Quantity
ds["Quantity"].fillna("-",inplace=True)
ds
```
